```python
def mean_eur2usd(x):
  return np.mean(x)*1.12
```

```python
df.groupby('sex')[['total_bill','prct_tip']].apply(mean_eur2usd)
```

```python
df.groupby(['time','sex'])[['total_bill','prct_tip']].apply(mean_eur2usd)
```

```python
df.groupby(['time','sex'])[['total_bill','prct_tip']].apply(lambda x: np.mean(x)*1.12)
```

```python
df.groupby(['time','sex'])[['total_bill','prct_tip']].aggregate([np.mean, np.max])
```

```python
df.groupby(['time','sex'])[['total_bill','prct_tip']].agg([np.mean, np.max])
```

```python
dict_agg = {'tip': [min, max], 'total_bill': [np.mean, mean_eur2usd]}
```

```python
df.groupby(['time','sex'])[['total_bill','tip']].agg(dict_agg)
```
