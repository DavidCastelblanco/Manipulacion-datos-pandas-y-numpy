```python
import pandas as pd
```

```python
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3']},
    index = ['y2','y3','a2','a3'])

df_right = pd.DataFrame(
    {'Z':['z2','z3','z4','z5'],
     'A':['a2','a3','a4','a5'],
     'Y':['y2','y3','y4','y5']},
    index = [2,3,4,5])
```

```python
df_left
```

```python
df_right
```

```python
pd.merge(df_left, df_right) #se juntan a traves de la columna en comun, los elementos que estan en ambos df
```

```python
pd.merge(df_left, df_right, how ='inner', on ='Y') # Es bueno especificar el tipo de union y porque parametro se esta uniendo
```

```python
pd.merge(df_left, df_right, how ='inner', left_on ='Mix', right_on ='Y') # Para hacer inner join de dos columnas diferentes
```

```python
pd.merge(df_left, df_right, how ='inner', on ='Y')
```

```python
pd.merge(df_left, df_right, how ='left', on ='Y')
```

```python
pd.merge(df_left, df_right, how ='right', on ='Y')
```

```python
pd.merge(df_left, df_right, how ='outer', on ='Y')
```

```python
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3'],
     'A':['a0','a1','a2','a3']},
    index = ['y2','y3','a2','a3'])
```

```python
pd.merge(df_left, df_right, how ='outer', on = ['Y', 'A'], suffixes=['_left', '_right'])
```
