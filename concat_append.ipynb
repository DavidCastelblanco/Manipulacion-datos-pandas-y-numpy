```python
s1 = pd.Series(x1[0], index=['a', 'b', 'c', 'd', 'e'])
s1
```

```python
s2 = pd.Series(x2[0], index=['c', 'd', 'e', 'f', 'g'])
s2
```

```python
pd.concat([s1, s2]) #por las filas
```

```python
pd.concat([s1, s2], axis=1) #por columnas compartiendo indices repetidos
```

```python
s1.reset_index()
```

```python
s1.reset_index(drop=True)
```

```python
pd.concat([s1.reset_index(drop=True), s2.reset_index(drop=True)], axis=1)
```

```python
df1 = pd.DataFrame(np.random.rand(3,2)*10, columns=['a', 'b'])
df1
```

```python
df2 = pd.DataFrame(np.random.rand(3,2)*-1, columns=['a', 'b'], index=[2,3,4])
df2
```

```python
pd.concat([df1, df2]) #por filas
```

```python
pd.concat([df1, df2], axis=1) #por columnas compartiendo indices repetidos
```

```python
pd.concat([df1, df2], axis=1, join='inner') # concatenacion solo de indices compartidos
```

```python
pd.concat([df1.reset_index(drop=True), df2.reset_index(drop=True)], axis=1)
```
